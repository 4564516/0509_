```python
re_tag = re.compile(r'<[^>]+>')

def rm_tags(text):
  return re_tag.sub('',text)
```

```python
def read_file(filetype):

 path = "/content/aclImdb/"

 file_list = []

 pos_path = path + filetype+"/pos/"

 for f in os.listdir(pos_path):

  file_list += [pos_path+f]




 neg_path = path + filetype+"/neg/"

 for f in os.listdir(neg_path):

  file_list += [neg_path+f]



 print('read:', filetype, 'files:', len(file_list))




 all_labels = ([1] * 12500 + [0] * 12500)

 all_texts =[]

 for fi in file_list:

  with open(fi,encoding='utf-8') as file_input:

   all_texts += [rm_tags(" ".join(file_input.readlines()))]

 return all_labels,all_texts
```

```python
y_test, x_test = read_file("test")
```

```python
y_train, x_train = read_file("train")
```

```python
token = Tokenizer(num_words=2000)
```

```python
token.fit_on_texts(x_train)
```

```python
print(token.word_index)
```

```python
text = ["今天 天氣 不錯 適合 出去 玩"]
token1 = Tokenizer(num_words=4)
token.fit_on_texts(text)
print(token1.word_index)
```

```python
x_train_seq = token.texts_to_sequences(x_train)
x_test_seq = token.texts_to_sequences(x_test)
```

```python
print(x_train[0],len(x_train[0]))
```

```python
print(len(x_train_seq[0]))
print(len(x_train[0]))
```

```python
for i in range(0,10):
  print(len(x_train_seq[i]))
```

```python
x_train = pad_sequences(x_train_seq, maxlen=300)
x_test = pad_sequences(x_test_seq, maxlen=300)
```

```python
print("brfore processing:",len(x_train_seq[0]))
print(x_train_seq[0])
```

```python
print("after processing:",len(x_train[0]))
print(x_train[0])
```

```python
model = Sequential()
model.add(Embedding(output_dim=32, input_dim=2000, input_length=300))
model.add(Flatten())
model.add(Dense(units=256,kernel_initializer="normal",activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))
model.summary()
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
train_history = model.fit(np.array(x_train),np.array(y_train),batch_size=100,epochs=10,validation_split=0.2)
```

```python
scores = model.evaluate(np.array(x_test),np.array(y_test))
```

```python
pre = model.predict(np.array(x_test))
```
